```python
import random
from collections import defaultdict
all_labels =['B-Actor', 'B-Animal', 'B-Athlete', 'B-Biological_Thing', 'B-Body_Of_Water', 'B-Body_Part', 'B-Car', 'B-Chemical_Thing', 'B-City', 'B-Color', 'B-Company', 'B-Country', 'B-County', 'B-Currency', 'B-Director', 'B-Disease', 'B-Drug', 'B-Educational_Degree', 'B-Educational_Institution', 'B-Engineer', 'B-Ethnicity', 'B-Film', 'B-Food', 'B-God', 'B-Government', 'B-Hospital', 'B-Law', 'B-Living_Thing', 'B-Military', 'B-Military_Conflict', 'B-Monarch', 'B-Mountain', 'B-Musician', 'B-Nationality', 'B-Political_Party', 'B-Politician', 'B-Sports_Event', 'B-Sports_League', 'B-Sports_Team', 'B-Time', 'B-Title', 'B-Weapon', 'I-Actor', 'I-Airport', 'I-Athlete', 'I-Biological_Thing', 'I-Car', 'I-Chemical_Thing', 'I-City', 'I-Company', 'I-Computer', 'I-Country', 'I-Date', 'I-Director', 'I-Educational_Degree', 'I-Educational_Department', 'I-Educational_Institution', 'I-Film', 'I-Game', 'I-Government_Agency', 'I-Hospital', 'I-Instrument', 'I-Law', 'I-Military', 'I-Monarch', 'I-Mountain', 'I-Music', 'I-Musician', 'I-Political_Party', 'I-Religious_Leader', 'I-Software', 'I-Time' 'I-Title', 'I-Tv_Channel', 'I-Weapon', 'I-Written_Work', 'OUTSIDE']
target_labels=all_labels
def create_support_query_sets(sentences, labels,target_labels, seed=None):
    if seed is not None:
        random.seed(seed)  # Set the seed for the random number generator

    support_set = []
    support_set_labels=[]
    query_set = []
    query_set_labels=[]
    label_count_support = defaultdict(int)
    label_count_query = defaultdict(int)

    for sentence, word_labels in zip(sentences, labels):
        for label in word_labels:
            if label in target_labels:
                if label_count_support[label] < 20:
                    # Randomly decide whether to add the word to the support set or the query set
                    if random.random() < 0.5:  # 80% probability for support set
                        if label_count_support[label] < 20:
                            support_set.append(sentence)
                            support_set_labels.append(word_labels)
                            label_count_support[label] += 1

                    else:
                        if label_count_query[label] < 5:
                            query_set.append(sentence)
                            query_set_labels.append(word_labels)
                            label_count_query[label] += 1

    return label_count_support,label_count_query,support_set,query_set,support_set_labels,query_set_labels

```

```python
seed_value = 42  # Choose any integer as the seed value
support_set_count, query_set_count,support_set,query_set,support_set_labels,query_set_labels = create_support_query_sets(S,M,all_labels ,seed=seed_value)
print(len(support_set_count))
print(len(query_set_count))
print(len(support_set))
print(len(query_set))
print(len(support_set_labels))
print(len(query_set_labels))
```

```python
# Prototype Generation -> Take mean
means_by_class = final_concatenated_features_s.groupby('label').mean()

means_by_class
```

```python
def euclidean_distance(feature_vector, centroid):
    return np.linalg.norm(feature_vector - centroid)
distances_df = pd.DataFrame()

for i,centroid in means_by_class.iterrows():
    dist=[]
    for j,feature in x_test.iterrows():
        distance= np.linalg.norm(feature - centroid) 
        dist.append(distance)
    distances_df[i]=dist
        
```

```python
distances_df['predictions']=distances_df.idxmin(axis=1)
distances_df['actual_label']=final_concatenated_features_q['label']
distances_df
```

```python
# Evaluation using prototypical network
from sklearn.metrics import accuracy_score,confusion_matrix

accuracy = accuracy_score(distances_df['actual_label'],distances_df['predictions'])

print("Accuracy Score:", accuracy)

conf_matrix = confusion_matrix(distances_df['actual_label'],distances_df['predictions'])

print('confusion matrix')
print(conf_matrix)

classes = label_encoder.classes_
labels = label_encoder.transform(classes)
```
