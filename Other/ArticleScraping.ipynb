```python
import wikipedia
import os
import warnings
import random
import re
wikipedia.set_lang("ar")
```

```python
# topics = ['Country', 'Places', 'Holiday', 'Photography', 'Beach', 'Sea', 'Ocean', 'Mountains', 'Roads', 'University', 'Job', 'Work', 'Time', 'Construction', 'Nursing', 'Medicine', 'School', 'Racing', 'Programming', 'Computer', 'Computer Science', 'Life', 'Rockets', 'Art', 'Science', 'Engineering', 'Animals', 'Boats', 'Airplanes', 'Internet', 'Social Media', 'Clothes', 'Games', 'Design', 'Hardware', 'Software', 'Business', 'Economics', 'Accounting', 'Houses', 'Movies', 'Music', 'Museum', 'News', 'Sports', 'Technology', 'Cars', 'Celebrity', 'World Cup', 'Fifa', 'Space', 'Food']
topics = ['Artificial intelligence', 'Robotics technology', 'Renewable energy', 'Job market trends', 'Buildings', 'Civil Engineering','Marine biology', 'Building Materials', 'Tourist Places', 'University Life', 'Environmental Engineering', 'Career growth' 'Construction Safety', 'Surveying and land development', 'Health and longevity']
searches = []

for topic in topics:
    searches += wikipedia.search(topic, results = 1000)
print(len(searches))
```

```python
print(len(topics))
random.shuffle(searches)
```

```python
articleSummaries = []
articleTitleNames = []
i = 0
j = 0
for search in searches:
    
    print(str(round(i/300 * 100, 1)) + "% Complete! Count = ", j, "/",len(searches))
    j += 1
    if i == 300:
        break
    
    try:
        tmp = wikipedia.summary(search)
        if len(tmp) >= 2000 and len(tmp) <= 4000:
            i += 1
            articleSummaries.append(tmp)
#             articleTitleNames.append(search)

    except Exception:
        pass
```

```python
print(len(articleSummaries))
```

```python
directory = "Articles"
num = 1771
if os.path.exists(directory) == False:
    os.mkdir(directory)
    os.chdir(directory)

for i in range(0,len(articleSummaries)):    
#     with open(directory + '/' + str(i) + articleTitleNames[i] + ".txt", 'w') as f:
    num += 1
    with open(directory + '/Article' + str(num) + ".txt", 'w') as f:
        f.write(articleSummaries[i])
```
